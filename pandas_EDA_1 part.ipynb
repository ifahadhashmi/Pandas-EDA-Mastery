# splitting str into multipe columns


```python
df = pd.DataFrame({"Name": ["Fahad Hashmi", "Ch Abuzar", "Ch Ali", "Sheikh Mubashir", "Shahid Baloch", "Muhammad Abdullah", "Mian Zohaib"],
                   "location": ["Hashmi colony, Jatoi", "Mohala Camboh, Jatoi", "Dr Khalid Pul, Jatoi","Dr Khalid Street, Jatoi", "Basti hooth Baloch, Jatoi", "Sohrab Canel, Jatoi", "Joiya Street, Jatoi"]})
```

```python
df.head(8)
```

```python
df.Name.str.split(" ").head(10)
```

```python
#splitting new columns
df.Name.str.split(" ", expand=True)
```

```python
df[["first_name", "last_name"]]=  df.Name.str.split(" ", expand= True)
```

```python
df.head()
```

```python
df.location.str.split(" ", expand= True)
```

```python
df["location"] = df["location"].astype(str)

df["City"] = df["location"].apply(lambda x: x.split()[-1] if isinstance(x, str) and len(x.split()) > 0 else "")
df["House Location"] = df["location"].apply(lambda x: " ".join(x.split()[:-1]) if isinstance(x, str) and len(x.split()) > 1 else "")
```

```python
df.head(8)
```

```python
df.head()
```

```python
df = df[["first_name", "last_name", "City", "House Location"]]
df
```
