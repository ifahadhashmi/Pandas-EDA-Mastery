# splitting str into multipe columns


```python
#Creating Dataframe
df = pd.DataFrame({"Name": ["Fahad Hashmi", "Ch Abuzar", "Ch Ali", "Sheikh Mubashir", "Shahid Baloch", "Muhammad Abdullah", "Mian Zohaib"],
                   "location": ["Hashmi colony, Jatoi", "Mohala Camboh, Jatoi", "Dr Khalid Pul, Jatoi","Dr Khalid Street, Jatoi", "Basti hooth Baloch, Jatoi", "Sohrab Canel, Jatoi", "Joiya Street, Jatoi"]})
```

```python
#printing first 8 values in dataset
df.head(8)
```

```python
# Split names by space and show first 10
df.Name.str.split(" ").head(10)
```

```python
#splitting new columns
df.Name.str.split(" ", expand=True)
```

```python
# Split 'Name' into 'first_name' and 'last_name' columns
df[["first_name", "last_name"]]=  df.Name.str.split(" ", expand= True)
```

```python
#printing first 5 values in Dataset
df.head()
```

```python
# Split 'location' column into separate columns using space as delimiter
df.location.str.split(" ", expand= True)
```

```python
# Convert 'location' column to string type to ensure consistency
df["location"] = df["location"].astype(str)

# Extract the last word from 'location' as 'City'
df["City"] = df["location"].apply(lambda x: x.split()[-1] if isinstance(x, str) and len(x.split()) > 0 else "")

#Extract all words except the last one as 'House Location'
df["House Location"] = df["location"].apply(lambda x: " ".join(x.split()[:-1]) if isinstance(x, str) and len(x.split()) > 1 else "")
```

```python
#Display first 5 rows of Dataset
df.head()
```

```python
# Keep only selected columns in the DataFrame
df = df[["first_name", "last_name", "City", "House Location"]]
df
```

# Aggregate by multiple groups

```python
#Import libraries
import pandas as pd
import seaborn as sns

#Import Dataset
df = sns.load_dataset("titanic")
df.head()
```

```python
# Count the number of non-null entries in each column grouped by 'sex'
df.groupby("sex").count()
```

```python
# Count the number of non-null entries in each column grouped by 'who'
df.groupby("who").count()
```

```python
# Get the total number of entries in the 'who' column
len(df.who)
```

```python
# Get the number of unique groups in the 'who' column
len(df.groupby("who"))
```

```python
# Count non-null entries in each column grouped by 'sex', 'pclass', and 'who'
df.groupby(["sex", "pclass", "who"]).count()
```

```python
# Categorize 'age' into bins: child, young_adult, and adult, and show first 5
pd.cut(df.age, bins = [0,18,25,99], labels=["child", "young_adult", "adult"]).head()
```
