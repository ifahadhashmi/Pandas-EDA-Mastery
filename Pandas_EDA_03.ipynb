# Lets download the titanic dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load the dataset
df= sns.load_dataset("titanic")
```

```python
df.head()
```

# Basic Information

```python
df.info()
```

```python
df.describe()
```

```python
len(df) #num of total values in column
```

```python
#Find missing values/ null values/NaNs
df.isnull().sum() #Count missing values
df.isnull().sum() /len(df) * 100 #percentage of missing values in a column
```

```python
#heatmap for displaying missing values
sns.heatmap(df.isnull(), cbar=False)
```

# Mean
mean is the average value of numeric series it helps in understanding the central tendency of dataset

# Median
Median is the middle value of sorted list of numbers.
If list has an odd number of values the middle number is the median
 or if list has an even number of values the average of two middle number is median

# Mode
Mode is the most frequently occuring value in dataset
A dataset can have more than one mode If all values are unique the mode will return all values as equally frequent
